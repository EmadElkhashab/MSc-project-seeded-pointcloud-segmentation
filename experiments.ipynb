```python
import os
import numpy as np
import open3d as o3d
from scipy.spatial import cKDTree

# Paths
pcd_path = "data/Stanford3dDataset_v1.2_Aligned_Version/Area_1/conferenceRoom_1/conferenceRoom_1.txt"
annotations_folder = "data/Stanford3dDataset_v1.2_Aligned_Version/Area_1/conferenceRoom_1/Annotations"

# Load full point cloud (XYZRGB)
point_cloud_data = np.loadtxt(pcd_path)
points = point_cloud_data[:, 0:3]
colors = point_cloud_data[:, 3:6] / 255.0  # normalize to [0,1]

# Create Open3D PointCloud
o3d_pcd = o3d.geometry.PointCloud()
o3d_pcd.points = o3d.utility.Vector3dVector(points)
o3d_pcd.colors = o3d.utility.Vector3dVector(colors)

# Assign random colors to annotation files
annotation_files = os.listdir(annotations_folder)
label_colors = {}

for ann_file in annotation_files:
    label = os.path.splitext(ann_file)[0]
    label_colors[label] = np.random.rand(3)  # random RGB

tree = cKDTree(points)

# Initialize semantic colors for all points
semantic_colors = np.zeros_like(points)

# For each annotation file, load its points and assign color
for ann_file in annotation_files:
    ann_path = os.path.join(annotations_folder, ann_file)
    ann_points = np.loadtxt(ann_path)[:, 0:3]  # XYZ
    label = os.path.splitext(ann_file)[0]
    color = label_colors[label]

    # Find nearest neighbors in full cloud
    dists, idxs = tree.query(ann_points, k=1)

    # Apply a tolerance (to avoid accidental mismatches)
    mask = dists < 1e-6
    semantic_colors[idxs[mask]] = color

# Update point cloud with semantic colors
o3d_pcd.colors = o3d.utility.Vector3dVector(semantic_colors)

# Visualize
o3d.visualization.draw_geometries([o3d_pcd])
```
